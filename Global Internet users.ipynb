```python
df_2018 = df[df['Year'] == 2018]

df_2018
```

```python
fig = px.choropleth(df_2018, 
                    locations='Code', 
                    color='No. of Internet Users', 
                    hover_name='Entity', 
                    color_continuous_scale='RdBu',
                    range_color=(0, 200000000),
                    )

fig.update_layout(
    title_text='Number of Internet Users in 2018',
    coloraxis_colorbar=dict(
        title='No. of Internet Users',
        dtick=20000000,
        tickformat='.0s',
    )
)

fig.show()
```

```python
fig = px.choropleth(df_2018, locations='Code', 
                    color='Internet Users(%)', 
                    hover_name='Entity', 
                    color_continuous_scale='RdBu')

fig.update_layout(title_text = 'Internet Users (%) in 2018')
fig.show()
```

```python
fig = px.choropleth(df_2018, 
                    locations='Code', 
                    color='Cellular Subscription', 
                    hover_name='Entity', 
                    color_continuous_scale='RdBu',
                    range_color=(0, 100)  # Set the color range from 0 to 100
                    )

fig.update_layout(
    title_text='Cellular Subscriptions per 100 people in 2018',
    coloraxis_colorbar=dict(
        title='Cellular Subscription',
        dtick=10
    )
)

fig.show()
```

```python
fig = px.choropleth(df_2018, locations='Code', 
                    color='Broadband Subscription', 
                    hover_name='Entity', 
                    color_continuous_scale='RdBu')

fig.update_layout(title_text = 'Broadband Subscriptions per 100 people in 2018')
fig.show()
```
